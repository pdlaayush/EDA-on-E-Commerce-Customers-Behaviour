## Library Importation
*import libraries for Data manipulation and visualization*

```python
#importing numpy pandas mathplotand other methods
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Data Inspection

```python
#first read the data file
df=pd.read_csv('E-commerce Customer Behavior - Sheet1.csv')
```

## Data Exploration

```python
# quick look to dataset
df.head(20)
```

```python
# Check shape
df.shape
```

```python
#Print summary statistics
df.describe()
```

```python
#the complete information about the dataset
df.info()
```

## Inspecting and Replacing missing values

```python
# Inspect missing values in the dataset
print(df.isnull().values.sum())

# Replace the ' 's with NaN
df = df.replace(" ",np.NaN)


# Impute the missing values with mean imputation
df = df.fillna(df.mean())

# Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

# Our EDA cycle starts from here.

## Checking the gender Distribution

```python
df['Gender'].hist(bins=20)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

* The frequency distribution for each gender is the same i.e. 175 each.

## Total Spend by Gender

```python
sns.boxplot(x='Gender', y='Total Spend', data=df)
plt.title('Total Spend By Gender')
plt.show()
```

* It seems that Male Users spend more on the Business. 

## Satisfaction Level

```python
sns.countplot(x=df['Satisfaction Level'])
```

* This bar graph shows the satisfaction level of E-commerece Customers according to the dataset where 122 peoples are satisfied whereas 110 are neutral and 118 are unsatisfied with E-commerce

## Satisfication Level According to Age

```python
sns.barplot(x=df['Satisfaction Level'],y=df['Age'])
```

## Satisfaction Level According to Items Purchased.

```python
sns.barplot(x=df['Satisfaction Level'],y=df['Items Purchased'])
```

## Pie-Chart Showing the Percentage of E-Commerce customers in Various Cities.

```python
city_df=df.groupby("City")
label_df=city_df['Items Purchased'].sum().sort_values(ascending=False).index
val_df=city_df['Items Purchased'].sum().sort_values(ascending=False).values
plt.pie(x=val_df,labels=label_df,autopct="%1.2f%%")
plt.show()
```

* People from 6 cities are involved in E-commerce.
* Maximum number of Customers are from San Franciso i.e. 26.30%.
* Minimum numbers of Customers are from Houston i.e. 9.95%.

## Heatmap of E-Commerce Customers Behaviour

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

##  Satisfaction Level of Customers according to Total Spending.

```python
new_df=df.groupby("Satisfaction Level")
label_df=new_df['Total Spend'].mean().index
val_df=new_df['Total Spend'].mean().values
plt.pie(x=val_df,labels=label_df,autopct="%1.2f%%")
plt.show()
```

 This Pie-Chart shows
* 51.38% are satisfied.
* 24.74  are neutral.
* 23.88 are unsatisfied.


## Total Spending According to Gender

```python
new_df=df.groupby("Gender")
label_df=new_df['Total Spend'].mean().index
val_df=new_df['Total Spend'].mean().values
plt.pie(x=val_df,labels=label_df,autopct="%1.2f%%")
plt.show()
```

This Pie-Chart shows the number of Male customers are more than Female.
i.e. 58.37 are Male and 41.63 are Female


## Total Spend VS Items Purchased in Scatter Plot

```python
plt.scatter(df['Total Spend'], df['Items Purchased'])
plt.xlabel('Total Spend')
plt.ylabel('Items Purchased')
plt.title('Scatter Plot: Total Spend vs. Items Purchased')
plt.show()
```

* This Scatter Plot shows the total spend vs Items Purchased by Customers using E-Commerce in the given datasets.

## Discount Applied

```python

df['Discount Applied'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title("Frequency of Discounts Applied")
plt.show()
```

* Frequency of Discount Applied is equal.

## Total Spend VS Items Purchased in Regression Plot

```python

sns.regplot(x='Total Spend', y='Items Purchased', data=df, scatter_kws={'s': 50, 'color': 'green'}, line_kws={'color': 'yellow'})

plt.xlabel('Total Spend')
plt.ylabel('Items Purchased')
plt.title('Regression Plot: Total Spend vs. Items Purchased')

plt.show()
```
